# Titanic Dataset - EDA

## Loading Titanic Dataset

```python
#Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="white")
```

```python
df = sns.load_dataset('titanic') #load an example dataset from the
df.head()
```

```python
print("Shape of the Dataset: ", df.shape) # number of rows and columns of a dataFrame
```

```python
df.info() #provides a concise summary of a dataframe
```

```python
df.isnull().sum() # returns the number of missing values in the dataset.
```

```python
sns.countplot(x='survived', data=df)
plt.title('Survival Count (0 = Died, 1 = Survived)')
plt.show()
```

```python
sns.countplot(x='sex', hue='survived',data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
sns.countplot(x='pclass', hue='survived', data=df)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.histplot(data=df, x='age', bins=30, kde=True)
plt.title('Passengers Age')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='survived', y='age', data=df)
plt.title('Survival by Age')
plt.ylabel('Age')
plt.show()
```

## Correlation Heatmap and Pairplot

```python
plt.figure(figsize=(11,7))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df[['pclass', 'fare', 'age', 'survived']], hue='survived')
plt.suptitle('Key features pairplot', y=1.03)
plt.show()
```

```python
sns.countplot(x='alone', hue='survived', data=df)
plt.show()
```
